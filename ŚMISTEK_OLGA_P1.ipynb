W dokumentacji dotyczącej danych czytamy:


> When computing daily counts, you may end up with negative numbers. This is a known issue due to decreasing cumulative counts in the original data provider.



Aby zabezpieczyć się przed tego rodzaju błędami, wszystkie zmienne skumulowane zostaną poddane operacji przypisania do błędnej obserwacji bieżącej - najbliższej, poprzedniej obserwacji poprawnej.

```python
cols = ['confirmed', 'deaths', 'tests', 'vaccines', 'people_vaccinated', 'people_fully_vaccinated']
```

```python
def cumulative(column):
  temp = column.to_numpy()
  for i in range(1, column.shape[0]):
    if temp[i] < temp[i-1]:
      temp[i] = temp[i-1]
    column = temp
```

```python
for col in cols:
  cumulative(data[col])
```

Na podstawie powyższej tabeli korelacji zmienne niezależne *population* oraz *contact_tracing* zostaną wyeliminowane, ponieważ nie są one w żaden sposób skorelowane ze zmiennymi niezależnymi *confirmed* oraz *deaths*.

Modele regresji liniowej zostaną wykonane osobno dla zmiennej niezależnej *confirmed* oraz *deaths*.

```python
def linear_regression(variable, X_train, X_test, y_train, y_test, result):
  model = LinearRegression()
  model = model.fit(X_train[[variable]], y_train)

  Y_pred_train = model.predict(X_train[[variable]])
  Y_pred_test = model.predict(X_test[[variable]])

  var = [model.score(X_train[[variable]], y_train), model.score(X_test[[variable]], y_test),
                      metrics.mean_squared_error(y_train, Y_pred_train, squared=False),
                      metrics.mean_squared_error(y_test, Y_pred_test, squared=False), 
                      metrics.mean_absolute_error(y_train, Y_pred_train),  metrics.mean_absolute_error(y_test, Y_pred_test)]

  result.loc[len(result)] = var
  return result
```

Modele regresji liniowej zostaną wykonane osobno dla zmiennej niezależnej *confirmed* oraz *deaths*.

```python
def linear_regression(variable, X_train, X_test, y_train, y_test, result):
  model = LinearRegression()
  model = model.fit(X_train[[variable]], y_train)

  Y_pred_train = model.predict(X_train[[variable]])
  Y_pred_test = model.predict(X_test[[variable]])

  var = [model.score(X_train[[variable]], y_train), model.score(X_test[[variable]], y_test),
                      metrics.mean_squared_error(y_train, Y_pred_train, squared=False),
                      metrics.mean_squared_error(y_test, Y_pred_test, squared=False), 
                      metrics.mean_absolute_error(y_train, Y_pred_train),  metrics.mean_absolute_error(y_test, Y_pred_test)]

  result.loc[len(result)] = var
  return result
```
